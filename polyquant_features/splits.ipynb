```python
import sqlite3
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

db_path = r"C:\Users\nimro\PolyQuant\data\sql\polymarket.db"
out_path = r"C:\Users\nimro\PolyQuant\data\features_v2\market_splits.parquet"

con = sqlite3.connect(db_path)
df = pd.read_sql_query(
    "SELECT condition_id AS market_id, effective_end_ts, split FROM market_splits;",
    con,
)
con.close()

pq.write_table(pa.Table.from_pandas(df), out_path)
print("wrote", out_path, "rows=", len(df))
```
